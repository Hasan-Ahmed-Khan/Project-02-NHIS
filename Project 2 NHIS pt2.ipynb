# Check for duplicate values

```python
df_b[df_b.duplicated()]
```

### No duplicates found

# Seperate numerical and categorical columns to perform data processing

```python
category_cols = ['season','yr','mnth','hr','weekday','weathersit']
```

```python
df_b[category_cols] = df_b[category_cols].astype('category')
```

```python
num_cols = ['temp','atemp','hum','windspeed','casual','registered','cnt']
```

```python
df_b.info()
```

# Find and handle Outliers

```python
# Compute IQR for each numeric column
Q1 = df_b[num_cols].quantile(0.25)
Q3 = df_b[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = (df_b[num_cols] < lower_bound) | (df_b[num_cols] > upper_bound)
outliers_sum = outliers.sum()  

# Replace outliers with the median of the column
for col in num_cols:
    median_value = df_b[col].median()
    df_b[col] = df_b[col].mask(outliers[col], median_value)

# Display outliers
outliers_sum
```

# Visualize the correlation matrix

```python
correlation_matrix = df_b[num_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Values')
plt.show()
```

```python
correlation_matrix
```

# Find skewness and kurtosis for numerical columns

```python
from scipy.stats import skew, kurtosis

skewness = df_b[num_cols].apply(skew)
kurt = df_b[num_cols].apply(kurtosis)

# Combine results into a dataframe
distribution_stats = pd.DataFrame({'Skewness': skewness, 'Kurtosis': kurt})
distribution_stats
```

# EDA on final cleaned dataset

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
fig.suptitle('Distribution of Numerical columns', fontsize=15)

for i, col in enumerate(num_cols):
    row, col_idx = divmod(i, 3)
    sns.histplot(df_b[col], bins=30, kde=True, ax=axes[row, col_idx])
    axes[row, col_idx].set_title(f'Distribution of {col}')
    
plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

```python
# Average bike rentals by weather 
plt.figure(figsize=(8, 4))
sns.barplot(x='weathersit', y='cnt', data=df_b, hue='weathersit', legend=False, palette='cividis')
plt.title('Average Bike Rentals by Weather Condition')
plt.xlabel('1. Clear / Few clouds / Partly cloudy \n2. Mist+Cloudy / Mist+Broken clouds / Mist+Few clouds / Mist \n3. Light Snow / Light Rain+Thunderstorm+Scattered clouds / Light Rain+Scattered clouds \n4. Heavy Rain+Ice Pallets+Thunderstorm+Mist / Snow+Fog ')
plt.ylabel('Average Rentals')
plt.show()
```

```python
# Average bike rentals on holidays vs. non-holidays
plt.figure(figsize=(6, 4))
sns.barplot(x='holiday', y='cnt', data=df_b, hue='holiday', legend=False, palette='pastel')
plt.title('Average Bike Rentals on Holidays vs. Non-Holidays')
plt.xlabel('Holiday or not')
plt.ylabel('Average Rentals')
plt.xticks([0, 1], ['Non-Holiday', 'Holiday'])
plt.show()
```

```python
# Grouped bar chart for casual vs. registered rentals based on weather conditions
plt.figure(figsize=(8, 5))
sns.barplot(x='weathersit', y='value', hue='variable', data=df_b.melt(id_vars=['weathersit'], value_vars=['casual', 'registered']), errorbar=None,  palette=['blue', 'red'])
plt.title('Casual vs. Registered Rentals Across Weather Conditions')
plt.xlabel('1. Clear / Few clouds / Partly cloudy \n2. Mist+Cloudy / Mist+Broken clouds / Mist+Few clouds / Mist \n3. Light Snow / Light Rain+Thunderstorm+Scattered clouds / Light Rain+Scattered clouds \n4. Heavy Rain+Ice Pallets+Thunderstorm+Mist / Snow+Fog ')
plt.ylabel('Average Rentals')
plt.legend(title='User Type', labels=['Casual', 'Registered'])
plt.show()
```
